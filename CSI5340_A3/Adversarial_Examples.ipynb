# IMPORTING DATA

```python
from torchvision import datasets
from torchvision.transforms import ToTensor

training_data = datasets.MNIST(
    root = 'data',
    train = True,                         
    transform = ToTensor(),
)

testing_data = datasets.MNIST(
    root = 'data', 
    train = False, 
    transform = ToTensor()
)
```

```python
BATCH_SIZE = 100

data_loaders = {
    'train' : DataLoader(training_data, batch_size=BATCH_SIZE, shuffle=True, num_workers=1),
    'test'  : DataLoader(testing_data, batch_size=BATCH_SIZE, shuffle=True, num_workers=1),
}
```

# PARAMETERS

```python
# CNN Training Params
LEARNING_RATE = 0.001
```

# HELPER FUNCTIONS

```python
def saveImage(data, index, model_name, true_label, perturbed_label, save=True):
    
    data = [item.detach().numpy() for item in data]
    image = data[index].reshape(28,28)
    true_label = true_label[index].numpy()
    perturbed_label = perturbed_label[index].numpy()
    
    fig, ax = plt.subplots(1)
    ax.imshow(image, cmap='gray')
    ax.text(2,2, 'PGD label='+str(perturbed_label), bbox={'facecolor': 'white', 'pad': 2}, fontsize=16)
    plt.axis('off')
    if save: plt.savefig("output/adv_imgs/"+model_name+"_"+str(true_label)+'.png')
```

# VISUALIZATION

```python
cnn_model = CNN()
optimizer_pgd = optim.Adam(cnn_model.parameters(), lr = LEARNING_RATE)
```

```python
for i, (training_data, training_label) in enumerate(data_loaders['train']):  
    if i in [0]:
        ex_train_img = training_data
        ex_train_lab = training_label
        break

(_, easy_images, easy_labels) = untargeted_pgd_attack(cnn_model, ex_train_img, ex_train_lab, eps=0.1, alpha=0.001, iterations=1)
(_, one_images, one_labels) = untargeted_pgd_attack(cnn_model, ex_train_img, ex_train_lab, eps=0.3, alpha=0.02, iterations=20)
(_, two_images, two_labels) = untargeted_pgd_attack(cnn_model, ex_train_img, ex_train_lab, eps=0.3, alpha=0.5, iterations=1)
(_, three_images, three_labels) = targeted_pgd_attack(cnn_model, ex_train_img, ex_train_lab, eps=0.3, alpha=0.02, iterations=20)
```

```python
saveImage(ex_train_img, 0, 'baseline', ex_train_lab, ex_train_lab)
saveImage(easy_images, 0, 'easyPerturb', ex_train_lab, easy_labels)
saveImage(one_images, 0, 'classifierOne', ex_train_lab, one_labels)
saveImage(two_images, 0, 'classifierTwo', ex_train_lab, two_labels)
saveImage(three_images, 0, 'classifierThree', ex_train_lab, three_labels)
```

# TESTING IMAGES

```python
for i, (testing_data, testing_label) in enumerate(data_loaders['test']):  
    if i in [0]:
        ex_test_img = testing_data
        ex_test_lab = testing_label
        break
        
for epsilon in [0, 0.1, 0.2, 0.3, 0.45]:

    (_, untarImg, untarLab) = untargeted_pgd_attack(cnn_model, ex_test_img, ex_test_lab, eps=epsilon, alpha=0.01, iterations=40)
    (_, tarImg, tarLab) = targeted_pgd_attack(cnn_model, ex_test_img, ex_test_lab, eps=epsilon, alpha=0.01, iterations=40)
    print(ex_test_lab[0])
    saveImage(untarImg, 0, 'test_untargeted_epsilon_'+str(epsilon), ex_test_lab, untarLab, save=True)
    saveImage(tarImg, 0, 'test_targeted_epsilon_'+str(epsilon), ex_test_lab, tarLab, save=True)
```
